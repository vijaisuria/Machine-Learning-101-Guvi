```python
import pandas as pd
import numpy as np
```

```python
# Download Dataset from here: htttps://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data
# Save it as adult.csv

original_data = pd.read_csv(
                    'data/adult.csv',
                     names = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", 
                              "occupation", "relationship", "race", "gender", "capital-gain", "capital-loss", "hours-per-week",
                              "country", "target"],
                     sep=r'\s*,\s*', 
                     engine='python', 
                     na_values='?'
                )

original_data.head()
```

```python
import sklearn.preprocessing as preprocessing

le = preprocessing.LabelEncoder()

original_data['target'] = le.fit_transform(original_data['target'])
original_data['occupation'] = le.fit_transform(original_data['occupation'])

print(original_data['occupation'].unique())
print(original_data['target'].unique().sort())

original_data.tail()
```
